```python
# import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
```

```python
from selenium.webdriver import ActionChains, Chrome
import time
```

```python
unit_trust = pd.read_csv('code_list.csv')
legend_list= pd.read_csv('legend_list.csv')
dict_Annualised_returns=pd.read_csv('ann_returns_columnheader.csv')
```

```python
url='https://investments3.personal-banking.hsbc.com.hk/public/utb/en-gb/fundDetail/'
```

```python

#ref_link=[]

#Fund profile

names=[]
latest_price=[]
fund_house=[]
currency=[]
Fund_class_inception_date=[]
Fund_size_Million_currency=[]
Fund_size_Million_amount=[]
Fund_size_Million_date=[]
HSBC_risk_level=[]
fiftytwoweek_range_currency=[]
fiftytwoweek_range_min=[]
fiftytwoweek_range_max=[]

#Investment strategy
HSBC_investment_category=[]
Investment_style=[]
Investment_instrument=[]

#Risk return profile(3years)
Annualised_return=[]


#Rating
Morningstar_rating_out_of_five=[]



#Fees and charges
HSBC_initial_charge=[]
Annual_management_fee_max=[]
HSBC_min_investment_amount_unit=[]
HSBC_min_investment_amount=[]

#Annualised returns
dict1_Annualised_returns= {'{}'.format(i):[] for i in dict_Annualised_returns['ann_returns_columnheader']}
#soup.find_all(class_="_16rlCwr0FAYME93uskRHIk")


dict_legend= {'{}'.format(i):[] for i in legend_list['all_legends']}
#soup.find_all("dd",attrs={"id":"legendValue"}).set()

```

```python
#test for loop
dict_legend['counter']=['0']

for i in range(len(unit_trust)):
#for i in range(len(unit_trust)):
    #get URL & append to list
    URL=url+unit_trust['Code'][i]
    try:
        try:
            driver = webdriver.Chrome()
            driver.maximize_window()
            driver.get(URL) #replacing the requests.get step
            time.sleep(3)
            driver.find_element_by_xpath("/html/body/div[1]/div/div[5]/div[3]/div[1]/div/div[2]/a/span").click()
        except:
            driver = webdriver.Chrome()
            driver.maximize_window()
            driver.get(URL) #replacing the requests.get step
            time.sleep(3)
            driver.find_element_by_xpath("/html/body/div[1]/div/div[5]/div[3]/div[1]/div/div[2]/a/span").click()
    except:
        try:
            driver = webdriver.Chrome()
            driver.maximize_window()
            driver.get(URL) #replacing the requests.get step
            time.sleep(3)
            driver.find_element_by_xpath("/html/body/div[1]/div/div[5]/div[3]/div[1]/div/div[2]/a/span").click()
        except:
            driver = webdriver.Chrome()
            driver.maximize_window()
            driver.get(URL) #replacing the requests.get step
            time.sleep(3)
            driver.find_element_by_xpath("/html/body/div[1]/div/div[5]/div[3]/div[1]/div/div[2]/a/span").click()
            
    #get URL source code

    subhtml = driver.page_source #replacing the html.text

    soup = BeautifulSoup(subhtml, "html.parser")

    soup.prettify()

    names.append(soup.find_all(class_="_3b-G7nRVvEHY34AYPke6Ap")[0].text)
    latest_price.append(soup.find_all(class_="_2BLgvf4O_8p7aXkqOJGrmO")[0].text.split()[1])
    fund_house.append(soup.find_all(class_="_29bc1orCEDBuJYpnVFEkwR")[0].text)
    currency.append(soup.find_all(class_="_29bc1orCEDBuJYpnVFEkwR")[2].text)
    Fund_class_inception_date.append(soup.find_all(class_="_29bc1orCEDBuJYpnVFEkwR")[1].text)
    Fund_size_Million_currency.append(re.search('[A-Za-z]+',soup.find_all(class_="_29bc1orCEDBuJYpnVFEkwR")[3].text).group())
    Fund_size_Million_amount.append(re.search('(\d*)(\,*)(\d+)(\.)(\d+)',soup.find_all(class_="_29bc1orCEDBuJYpnVFEkwR")[3].text).group())
    Fund_size_Million_date.append(soup.find_all(class_="_29bc1orCEDBuJYpnVFEkwR")[3].text[-11:])
    HSBC_risk_level.append(soup.find_all(class_="_29bc1orCEDBuJYpnVFEkwR")[4].text)
    try:
        fiftytwoweek_range_currency.append(re.search('[A-Za-z]+',soup.find_all(class_="_29bc1orCEDBuJYpnVFEkwR")[6].text).group())
    except:
        fiftytwoweek_range_currency.append(np.nan)
    try:
        fiftytwoweek_range_min.append(re.search('(\d*)(\,*)(\d+)(\.)(\d+)',soup.find_all(class_="_29bc1orCEDBuJYpnVFEkwR")[6].text).group())
    except:
        fiftytwoweek_range_min.append(np.nan)
    try:
        fiftytwoweek_range_max.append(re.search('(\d*)(\,*)(\d+)(\.)(\d+)(?!.*(\d*)(\,*)(\d+)(\.)(\d+))',soup.find_all(class_="_29bc1orCEDBuJYpnVFEkwR")[6].text).group())
    except:
        fiftytwoweek_range_max.append(np.nan)

    #Investment strategy(錯,要改,頭四已經唔同)
    HSBC_investment_category.append(soup.find_all(class_="_29bc1orCEDBuJYpnVFEkwR")[8].text)
    Investment_style.append(soup.find_all(class_="_29bc1orCEDBuJYpnVFEkwR")[9].text)
    Investment_instrument.append(soup.find_all(class_="_29bc1orCEDBuJYpnVFEkwR")[10].text)


    #Risk return profile(3years)
    Annualised_return.append(soup.find_all(class_="_29bc1orCEDBuJYpnVFEkwR")[-17].text)

    #Annualised_returns
    try:
        dict1_Annualised_returns['1 month'].append(re.findall('\+*\-*\d+\.+\d*\%',soup.find_all(class_="_1402b4AVBTNOhBcp9unB8U _3NqUuvbQB8nopIbw347iu1")[0].text)[0])
    except:
        dict1_Annualised_returns['1 month'].append(np.nan)
    try:
        dict1_Annualised_returns['3 month'].append(re.findall('\+*\-*\d+\.+\d*\%',soup.find_all(class_="_1402b4AVBTNOhBcp9unB8U _3NqUuvbQB8nopIbw347iu1")[0].text)[1])
    except:
        dict1_Annualised_returns['3 month'].append(np.nan)
    try:    
        dict1_Annualised_returns['6 month'].append(re.findall('\+*\-*\d+\.+\d*\%',soup.find_all(class_="_1402b4AVBTNOhBcp9unB8U _3NqUuvbQB8nopIbw347iu1")[0].text)[2])
    except:
        dict1_Annualised_returns['6 month'].append(np.nan)
    try:
        dict1_Annualised_returns['1 year'].append(re.findall('\+*\-*\d+\.+\d*\%',soup.find_all(class_="_1402b4AVBTNOhBcp9unB8U _3NqUuvbQB8nopIbw347iu1")[0].text)[3])
    except:
        dict1_Annualised_returns['1 year'].append(np.nan)
    try:
        dict1_Annualised_returns['3 year'].append(re.findall('\+*\-*\d+\.+\d*\%',soup.find_all(class_="_1402b4AVBTNOhBcp9unB8U _3NqUuvbQB8nopIbw347iu1")[0].text)[4])
    except:
        dict1_Annualised_returns['3 year'].append(np.nan)
    try:
        dict1_Annualised_returns['5 year'].append(re.findall('\+*\-*\d+\.+\d*\%',soup.find_all(class_="_1402b4AVBTNOhBcp9unB8U _3NqUuvbQB8nopIbw347iu1")[0].text)[5])
    except:
        dict1_Annualised_returns['5 year'].append(np.nan)
    try:
        dict1_Annualised_returns['10 year'].append(re.findall('\+*\-*\d+\.+\d*\%',soup.find_all(class_="_1402b4AVBTNOhBcp9unB8U _3NqUuvbQB8nopIbw347iu1")[0].text)[6])
    except:
        dict1_Annualised_returns['10 year'].append(np.nan)

    #Rating
    try:
        Morningstar_rating_out_of_five.append(soup.find_all(class_="_30mCakbI36jb_7H3dqRDdp")[0].text[0:2])
    except:
        Morningstar_rating_out_of_five.append(np.nan)

    #Fees and charges
    HSBC_initial_charge.append(soup.find_all(class_="_29bc1orCEDBuJYpnVFEkwR")[-8].text)
    Annual_management_fee_max.append(soup.find_all(class_="_29bc1orCEDBuJYpnVFEkwR")[-7].text)
    HSBC_min_investment_amount_unit.append(re.search('[A-Za-z]+',soup.find_all(class_="_29bc1orCEDBuJYpnVFEkwR")[-6].text).group())
    HSBC_min_investment_amount.append(re.search('(\d*)(\,*)(\d+)(\.*)(\d*)',soup.find_all(class_="_29bc1orCEDBuJYpnVFEkwR")[-6].text).group())

    #legend_value
    legend_val_list=list(set(soup.find_all(class_="_3vHpGqFaBI90QjB5BFFCiv")))
    for j in dict_legend.keys():
        for i in legend_val_list:
            if i.text.strip(re.search('(\d+)\.*(\d*)%',i.text).group()) ==j:
                dict_legend[i.text.strip(re.search('(\d+)\.*(\d*)%',i.text).group())].append(re.search('(\d+)\.*(\d*)%',i.text).group())
                break
        if len(dict_legend[j])<len(dict_legend['counter']):
                dict_legend[j].append('0%')
    
    dict_legend['counter'].append('0')
       
```
