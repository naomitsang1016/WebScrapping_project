```python
#fill missing data as 0
df['Annualised_return'] = df['Annualised_return'].replace("-","0")
```

```python
#remove percent(%) and change to demical numbers
def remove_percent(x):
    return round(float(x.replace("%",""))/100,4)
```

```python
df['Annualised_return'] = df['Annualised_return'].apply(remove_percent)
df['HSBC_initial_charge'] = df['HSBC_initial_charge'].apply(remove_percent)
df['Annual_management_fee_max'] = df['Annual_management_fee_max'].apply(remove_percent)

df['Asset allocation: Stock'] = df['Asset allocation: Stock'].apply(remove_percent)
df['Asset allocation: Bond'] = df['Asset allocation: Bond'].apply(remove_percent)
df['Asset allocation: Cash'] = df['Asset allocation: Cash'].apply(remove_percent)
df['Asset allocation: Cash & Equivalents'] = df['Asset allocation: Cash & Equivalents'].apply(remove_percent)
df['Asset allocation: Derivative'] = df['Asset allocation: Derivative'].apply(remove_percent)
df['Asset allocation: Others'] = df['Asset allocation: Others'].apply(remove_percent)

df['Financial Services'] = df['Financial Services'].apply(remove_percent)
df['Real Estate'] = df['Real Estate'].apply(remove_percent)
df['Corporate'] = df['Corporate'].apply(remove_percent)
df['Healthcare'] = df['Healthcare'].apply(remove_percent)
df['Utilities'] = df['Utilities'].apply(remove_percent)
df['Basic Materials'] = df['Basic Materials'].apply(remove_percent)
df['Communication Services'] = df['Communication Services'].apply(remove_percent)
df['Energy'] = df['Energy'].apply(remove_percent)
df['Consumer Cyclical'] = df['Consumer Cyclical'].apply(remove_percent)
df['Industrials'] = df['Industrials'].apply(remove_percent)
df['Securitized'] = df['Securitized'].apply(remove_percent)
df['Government'] = df['Government'].apply(remove_percent)
df['Consumer Defensive'] = df['Consumer Defensive'].apply(remove_percent)
df['Technology'] = df['Technology'].apply(remove_percent)

df['United States'] = df['United States'].apply(remove_percent)
df['Latin America'] = df['Latin America'].apply(remove_percent)
df['Canada'] = df['Canada'].apply(remove_percent)
df['United Kingdom'] = df['United Kingdom'].apply(remove_percent)
df['Europe - Emerging'] = df['Europe - Emerging'].apply(remove_percent)
df['Europe - ex Euro'] = df['Europe - ex Euro'].apply(remove_percent)
df['Eurozone'] = df['Eurozone'].apply(remove_percent)
df['Australasia'] = df['Australasia'].apply(remove_percent)
df['Asia - Developed'] = df['Asia - Developed'].apply(remove_percent)
df['Asia - Emerging'] = df['Asia - Emerging'].apply(remove_percent)
df['Middle East'] = df['Middle East'].apply(remove_percent)
df['Japan'] = df['Japan'].apply(remove_percent)
df['Africa'] = df['Africa'].apply(remove_percent)
```

```python
#Remove comma(,) between numbers
def remove_comma(x):
    return round(float(x.replace(",","")),4)
```

```python
df['Fund_size_Million_amount'] = df['Fund_size_Million_amount'].apply(remove_comma)
df['HSBC_min_investment_amount'] = df['HSBC_min_investment_amount'].apply(remove_comma)
```

```python
# Annualised_returns
def return_clean(x):
    return round(x.str.extract('([\+\-]\d+\.\d+)'),4)
```

```python
df['Annualised_returns: 1 month'] = return_clean(df['Annualised_returns: 1 month']).astype(float)/100
df['Annualised_returns: 3 month'] = return_clean(df['Annualised_returns: 3 month']).astype(float)/100
df['Annualised_returns: 6 month'] = return_clean(df['Annualised_returns: 6 month']).astype(float)/100
df['Annualised_returns: 1 year'] = return_clean(df['Annualised_returns: 1 year']).astype(float)/100
df['Annualised_returns: 3 year'] = return_clean(df['Annualised_returns: 3 year']).astype(float)/100
df['Annualised_returns: 5 year'] = return_clean(df['Annualised_returns: 5 year']).astype(float)/100
df['Annualised_returns: 10 year'] = return_clean(df['Annualised_returns: 10 year']).astype(float)/100
```

```python
#to_date
df['Fund_class_inception_date'] = pd.to_datetime(df['Fund_class_inception_date'])
df['Fund_size_Million_date'] = pd.to_datetime(df['Fund_size_Million_date'])
```

```python
#add column of min invest with USD as currency
df['HSBC_min_investment_amount_unit'].unique()

def convert_usd(x,y):
    #currency on 27/1/2022
    if y =='USD':
        return x
    elif y=='EUR':
        return x*1.12
    elif y=='CNY':
        return x*0.16
    elif y=='HKD':
        return x*0.13
    elif y =='SGD':
        return x*0.74
    elif y =='AUD':
        return x*0.71
    elif y=='NZD':
        return x*0.66
    elif y =='GBP':
        return x*1.34
    elif y =='CAD':
        return x*0.79    
    elif y =='JPY':
        return x*0.0087
    
df['HSBC_min_investment_amount_USD']=df.apply(lambda x: convert_usd(x.HSBC_min_investment_amount, x.HSBC_min_investment_amount_unit), axis=1)
```

```python
#change datatype of risk level to str
df['HSBC_risk_level'] = df['HSBC_risk_level'].astype(str)
```

```python
def country_max(x,a,b,c,d,e,f,g,h,i,j,k,l):
    
    Country = ['United States', 'Latin America', 'Canada', 'United Kingdom',
       'Europe - Emerging', 'Europe - ex Euro', 'Eurozone', 'Australasia',
       'Asia - Developed', 'Asia - Emerging', 'Middle East', 'Japan',
       'Africa']
    y=[x,a,b,c,d,e,f,g,h,i,j,k,l]
    z = max(y)
    for i in range(len(y)):
        if y[i] == z:
            return Country[i]
                
df['Equity_holdings_country_max']=df.apply(lambda x: country_max(x['United States'], x['Latin America'],x['Canada'],x['United Kingdom'],x['Europe - Emerging'],x['Europe - ex Euro'],x['Eurozone'],x['Australasia'],
x['Asia - Developed'],x['Asia - Emerging'],x['Middle East'],x['Japan'],x['Africa']), axis=1)
```

```python
def Equity_holdings_sector_max(x,a,b,c,d,e,f,g,h,i,j,k,l,m):
    
    sector = ['Financial Services', 'Real Estate',
       'Corporate', 'Healthcare', 'Utilities', 'Basic Materials',
       'Communication Services', 'Energy', 'Consumer Cyclical', 'Industrials',
       'Securitized', 'Government', 'Consumer Defensive', 'Technology']
    y=[x,a,b,c,d,e,f,g,h,i,j,k,l,m]
    z = max(y)
    for i in range(len(y)):
        if y[i] == z:
            return sector[i]
                
df['Equity_holdings_sector_max']=df.apply(lambda x: Equity_holdings_sector_max(x['Financial Services'], x['Real Estate'],x['Corporate'],x['Healthcare'],x['Utilities'],x['Basic Materials'],x['Communication Services'],x['Energy'],x['Consumer Cyclical'],x['Industrials'],
x['Securitized'],x['Government'],x['Consumer Defensive'],x['Technology']), axis=1)
```
