```python
plt.figure(figsize=(10,10))
a = df.groupby('HSBC_risk_level')['Annualised_returns: 1 month'].mean()*100
b = df.groupby('HSBC_risk_level')['Annualised_returns: 1 month'].max()*100
c = df.groupby('HSBC_risk_level')['Annualised_returns: 3 month'].mean()*100
d = df.groupby('HSBC_risk_level')['Annualised_returns: 3 month'].max()*100
e = df.groupby('HSBC_risk_level')['Annualised_returns: 6 month'].mean()*100
f = df.groupby('HSBC_risk_level')['Annualised_returns: 6 month'].max()*100
g = df.groupby('HSBC_risk_level')['Annualised_returns: 1 year'].mean()*100
h = df.groupby('HSBC_risk_level')['Annualised_returns: 1 year'].max()*100
plt.plot(a, color = 'forestgreen', linestyle = '--', label = '1mth mean')
plt.plot(b, color = 'forestgreen', linestyle = 'solid', label = '1mth max')
plt.plot(c, color = 'dodgerblue',linestyle = '--', label = '3mth mean')
plt.plot(d, color = 'dodgerblue', linestyle = 'solid', label = '3mth max')
plt.plot(e, color = 'tomato', linestyle = '--', label = '6mth mean')
plt.plot(f, color = 'tomato', linestyle = 'solid', label = '6mth max')
plt.plot(g, color = 'darkorchid', linestyle = '--', label = '1yr mean')
plt.plot(h, color = 'darkorchid', linestyle = 'solid', label = '1yr max')
plt.title("Short-term Investment Performance (Annualised Return) across Risk Level", fontsize = 15)  
plt.xlabel("Risk Level", fontsize = 15)  
plt.ylabel("Return(%)", fontsize = 15)  
plt.legend(loc = 'upper left', fontsize = 15) 
plt.savefig("Short-term Investment Performance (Annualised Return) across Risk Level", bbox_inches='tight')
plt.show()
```

```python
import numpy as np

font = {'family' : 'DejaVu Sans',
        'weight' : 'bold',
        'size'   : 70}

plt.rc('font', **font)

ind_frame1 = ind_frame.groupby('Period').sum()

x_ticks_labels = ['1month','3month','6month','1year','3year','5year','10year']

ind_frame1 = ind_frame1.reindex(x_ticks_labels, axis=0)

ax = ind_frame1.plot.barh(stacked=True,colormap='Reds')

plt.legend(bbox_to_anchor=(1.1, 1.05))
ax.set_xlabel("No. of Sector",fontsize = 80)
ax.set_ylabel("Period",fontsize = 80)
plt.title("No. of Sector within Top 50 across Period",fontsize = 120)

ind_frame_total = ind_frame["Financial Services"] + ind_frame["Technology"] + ind_frame["Corporate"] + ind_frame["Basic Materials"] + ind_frame["Energy"] + ind_frame["Consumer Cyclical"] + ind_frame["Industrials"] + ind_frame["Consumer Defensive"] + ind_frame["Government"] + ind_frame["Real Estate"] + ind_frame["Healthcare"]
ind_frame_rel = ind_frame.div(ind_frame_total, 0)*100

for n in ind_frame_rel:
    for i, (cs, ab, pc) in enumerate(zip(ind_frame.iloc[:,:].cumsum(1)[n],ind_frame[n], ind_frame_rel[n])):
        if pc != 0 :
            plt.text(cs - ab / 2, i, str(np.round(pc, 1)) + '%',va = 'center', ha = 'center')
plt.savefig("No. of Sector within Top 50 across Period.png", bbox_inches='tight')
plt.show()
```
